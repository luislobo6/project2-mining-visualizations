```python
import geopandas as gpd
```

```python
# Start reading one entity to start populate the DataFrame
mapa = gpd.read_file(f"conjunto_de_datos/01ent.shp")

# with a for cicle starting from the second entity to the 32th, read and append to the DataFrame
for x in range(2, 33):
    if (x < 10):
        mapa = mapa.append(gpd.read_file(f"conjunto_de_datos/0{x}ent.shp"))
    else:
        mapa = mapa.append(gpd.read_file(f"conjunto_de_datos/{x}ent.shp"))
```

```python
# check the DataFrame
mapa
```

```python
# informative
mapa.crs
```

```python
# transform from the inegi type of coordinates to the cartesian regular ones
mapa = mapa.to_crs("EPSG:4326")
```

```python
# plot the map with all the entities
mapa.plot()
```

```python
# save into GeoJSON type of file to use in the front end 
# the coordinates are backwards, when used first use (coordinates[1],coordinates[0])
mapa.to_file("mexico_ent.geojson", driver="GeoJSON")
```
